### **실습1**
1. purchase_date 시계열 데이터로 변경
2. item_name 컬럼의 값들을 모두 대문자로 변경
3. item_name 컬럼의 값들의 공백 모두 제거
4. item_name을 기준으로 오름차순 정렬
5. 인덱스 재지정

```python
import pandas as pd
uriage = pd.read_csv('./datasets/uriage.csv')
uriage
```

```python
# 1. purchase_date 시계열 데이터로 변경
uriage['purchase_date'] = pd.to_datetime(uriage['purchase_date'])
uriage.info()
```

```python
# 2. item_name 컬럼의 값들을 모두 대문자로 변경
uriage['item_name'] = uriage['item_name'].str.upper()
uriage
```

```python
# 3. item_name 컬럼의 값들의 공백 모두 제거
uriage['item_name'] = uriage['item_name'].str.replace(' ', '')
uriage
```

```python
# 4. item_name을 기준으로 오름차순 정렬
uriage.sort_values('item_name', ascending=True, inplace=True)
uriage
```

```python
# 5. 인덱스 재지정
uriage.reset_index(drop=True, inplace=True)
uriage
```

### **실습2**
1. transation_1, transaction_2 데이터 프레임 단순항 행 결합
2. 결합한 데이터에서 price가 100000 미만인 행 제거
3. 1, 2번 처리 후 고유한 customer_id는 몇개인지 구하여라

```python
transaction_1 = pd.read_csv('./datasets/transaction_1.csv')
transaction_2 = pd.read_csv('./datasets/transaction_2.csv')
```

```python
transaction_1.head()
```

```python
transaction_2.head()
```

```python
# 1. transation_1, transaction_2 데이터 프레임 단순항 행 결합
transaction = pd.concat([transaction_1, transaction_2], axis=0)
transaction
```

```python
transaction.reset_index(drop=True, inplace=True)
```

```python
# 2. 결합한 데이터에서 price가 100000 미만인 행 제거
idx = transaction[transaction['price']<100000].index
tran = transaction.drop(idx, axis=0)
tran
```

```python
# 3. 1, 2번 처리 후 고유한 customer_id는 몇개인지 구하여라
tran['customer_id'].nunique()
```

### **실습3**
1. 'Order ID', 'Total Revenue', 'Total Cost', 'Total Profit' 컬럼 삭제
2. 컬럼 이름 변경 -> ['권역', '국가', '상품종류', '채널', '우선순위', '주문일자', '방송일자', '판매단위', '단가', '원가']
3. '단가'*'판매단위' -> '총 단가' 파생변수 생성
4. '권역', '국가'를 기준으로 오름차순 정렬
5. index 값 초기화하고 기존의 인덱스는 삭제

```python
sales = pd.read_csv('./datasets/Sales Records.csv')
sales.head()
```

```python
# 1. 'Order ID', 'Total Revenue', 'Total Cost', 'Total Profit' 컬럼 삭제
sales.drop(['Order ID', 'Total Revenue', 'Total Cost', 'Total Profit'], axis=1, inplace=True)
sales.head()
```

```python
# 2. 컬럼 이름 변경 -> ['권역', '국가', '상품종류', '채널', '우선순위', '주문일자', '방송일자', '판매단위', '단가', '원가']
sales.columns = ['권역', '국가', '상품종류', '채널', '우선순위', '주문일자', '방송일자', '판매단위', '단가', '원가']
sales.head()
```

```python
# 3. '단가'*'판매단위' -> '총 단가' 파생변수 생성
sales['총 단가'] = sales['단가'] * sales['판매단위']
sales.head()
```

```python
# 4. '권역', '국가'를 기준으로 오름차순 정렬
sales.sort_values(['권역', '국가'], ascending=True, inplace=True)
sales.head()
```

```python
# 5. index 값 초기화하고 기존의 인덱스는 삭제
sales.reset_index(drop=True, inplace=True)
sales.head()
```
