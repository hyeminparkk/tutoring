### **1. 변수 이름 변경 / 변수 추가 & 삭제**

```python
import pandas as pd
df = pd.DataFrame({'ID' : [1011, 1012, 1013, 1014, 1015],
                    'name' : ['Kim', 'Lee', 'Jeong', 'Park', 'Song'],
                    'age' : [20, 19, 36, 24, 15]})
df
```

#### **1.1 변수명 변경**

```python
# ID -> Studen_id로 변수명 변경
df.rename(columns={'ID':'Student_id'}, inplace=True)
df
```

```python
df.columns = ['아이디', '성', '나이']
```

```python
df
```

#### **1.2 파생변수 생성**

#### **iloc와 loc 차이**
- iloc : 인덱스 번호 사용하는 것 (iloc[행 번호, 열 번호])
- loc : 변수명 사용 (loc[행 번호, 컬럼명])

```python
# 파생변수 생성 -> age 변수를 이용한 age_group 변수 생성
# loc 사용
df['age_group'] = 0
for i in range(0, len(df)):
    if df.loc[i, 'age'] < 20 :
        df.loc[i, 'age_group'] = '10대'
    elif df.loc[i, 'age'] < 30 :
        df.loc[i, 'age_group'] = '20대'
    elif df.loc[i, 'age'] < 40 :
        df.loc[i, 'age_group'] = '30대'
df
```

```python
# 파생변수 생성 -> age 변수를 이용한 age_group 변수 생성
# iloc 사용
df['age_group'] = 0
for i in range(0, len(df)):
    if df.iloc[i, 2] < 20 :
        df.iloc[i, 3] = '10대'
    elif df.iloc[i, 2] < 30 :
        df.iloc[i, 3] = '20대'
    elif df.iloc[i, 2] < 40 :
        df.iloc[i, 3] = '30대'
df
```

#### **1.3 변수 삭제**

```python
del df['age_group']
df
```

```python
df.drop(['age'], axis=1, inplace=True)
df
```

### **2. 중복행 처리**

```python
df = pd.DataFrame({'ID' : [1011, 1012, 1013, 1014, 1015, 1014],
                    'name' : ['Kim', 'Kim', 'Jeong', 'Park', 'Song', 'Park'],
                    'age' : [20, 19, 36, 24, 15, 24]})
df
```

```python
# 중복행 확인
df.duplicated()
```

```python
sum(df.duplicated())
```

```python
# 한 열에서의 중복 확인
df.duplicated('name')
```

#### **drop_duplicates()**
- keep = 'first' - 처음 행만 남김 (기본값)
- keep = 'last' - 마지막 행만 남김
- keep = False - 중복된 행 모두 제거

```python
# 중복행 제거
df.drop_duplicates(keep='first', inplace=True)
df
```

```python
# 중복행 제거 후 reset_index()
df.reset_index(drop=True, inplace=True)
df
```

```python
# 특정 열들이 중복된 경우 행 제거
df = pd.DataFrame({'주문날짜' : ['2020-01-01', '2020-01-01', '2020-01-01', '2020-01-02', '2020-01-02', '2020-01-05'],
                  '상품번호' : ['T1001', 'S1001', 'T1001', 'S1001', 'S1001', 'T1003'], 
                  '가격' : [1000, 2000, 1000, 2000, 2000, 3000]})
df
```

### **4. 데이터 병합 - merge/concat**

#### **merge**

```python
# 주문정보 df
df_1 = pd.DataFrame({'ID' : [1, 2, 3, 4, 5],
                         'Name' : ['Kim', 'Park', 'Song', 'Kim', 'Jeong']})
df_2 = pd.DataFrame({'ID' : [2, 3, 5, 6], 
                     'Address' : ['Seoul', 'Dague', 'Busan', 'Seoul']})
```

```python
df_1
```

```python
df_2
```

```python
df_3 = pd.merge(df_1, df_2, on='ID', how='inner')
df_3
```

```python
df_4 = pd.merge(df_1, df_2, on='ID', how='left')
df_4
```

```python
df_5 = pd.merge(df_1, df_2, on='ID', how='right')
df_5
```

```python
df_6 = pd.merge(df_1, df_2, on='ID', how='outer')
df_6
```

#### **concat**

```python
df = pd.DataFrame({'S1': [1,2,3,4], 
                   'S2': [5,6,7,8]})
df
```

```python
df3 = pd.DataFrame({'S1': [1,3]})
df3
```

```python
pd.concat([df, df3], axis=0)
```

```python
pd.concat([df, df3], axis=1)
```

### **8. 기본 함수**

```python
# 대소문자 변환
a = 'Hello World'
print(a.upper()) # 대문자로
print(a.lower()) # 소문자로
print(a.title()) # 단어의 첫 글자만 대문자
print(a.swapcase()) # 소문자는 대문자로, 대문자는 소문자로
```

```python
# 공백 제거 - strip(), replace()
a = '    hello world  '
print(a.lstrip())
print(a.rstrip())
print(a.strip())

print(a.replace(" ", ""))
```

```python
# 문자열 나눠주는 함수 (리스트 변환) - split()
a = 'a,b,c,d,e'
a.split(',')
```
